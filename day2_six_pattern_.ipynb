2. Dependency Injection (DI)
Most asked interview topic.
________________________________________
What is Dependency?
Example:
class Car {
    Engine engine = new Engine();
}
Car depends on Engine.
Engine is a dependency.
________________________________________
Problem
Car creates Engine tightly.
Hard to:
•	test
•	modify
•	replace
________________________________________
Spring Solution
Inject dependency from outside.
class Car {
    Engine engine;

    Car(Engine engine){
        this.engine = engine;
    }
}
Now Spring supplies Engine automatically.
This is Dependency Injection.
________________________________________

1. Constructor Injection

}
________________________________________
2. Setter Injection

public void setEngine(Engine engine){
   this.engine = engine;
}

Inversion of Control (IOC)
Inversion of Control is the core idea behind Spring Framework. Normally in Java, programmers create objects manually using new, but in Spring, the framework creates and manages objects automatically. This “control” shifts from developer to Spring container, which is why it is called IOC. In interviews, explain it simply as: “Spring manages object creation and lifecycle instead of the programmer managing it manually.”
________________________________________
Dependency Injection (DI)
Dependency Injection is a feature built on top of IOC. Instead of a class creating its own dependencies, Spring injects them automatically. For example, if a StudentService needs a StudentRepository, Spring connects them using @Autowired. This reduces tight coupling and makes code easier to test and maintain. Interviewers often ask DI because it shows whether you understand real-world backend architecture and clean coding practices.
________________________________________
Spring Beans
A Bean is simply an object managed by the Spring container. When you use annotations like @Component, @Service, or @Repository, Spring automatically creates those objects and stores them as beans. Instead of manually creating objects with new, developers let Spring manage them. In placements, students usually get confused between normal Java objects and Spring Beans, so remember: every bean is an object, but every object is not necessarily a Spring Bean.
________________________________________
Spring Boot
Spring Boot is an extension of Spring that simplifies application development by reducing configuration. Earlier, Spring applications required large XML configurations and external servers, but Spring Boot provides auto-configuration and embedded servers like Tomcat. This helps developers quickly build REST APIs and microservices. Most companies today use Spring Boot, so interviewers expect students to know why it is faster and easier than traditional Spring.
________________________________________
Spring MVC
Spring MVC is a web framework inside Spring used to build web applications and REST APIs using the Model-View-Controller pattern. The Controller handles requests, the Model manages data, and the View displays output to users. For example, when a user opens /students, the controller processes the request, fetches data from service classes, and returns the response. In placements, students should clearly understand the MVC flow because it is one of the most common backend architecture questions.
